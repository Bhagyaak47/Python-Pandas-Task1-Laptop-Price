```python
import pandas as pd
a=pd.read_csv("laptop_price .csv")

df=pd.DataFrame(a)

df["8GB"]=df["Ram"].apply(lambda x: "8GB" if x=="8GB" else "NA")
df["4GB"]=df["Ram"].apply(lambda x: "4GB" if x=="4GB" else "NA")
df["16GB"]=df["Ram"].apply(lambda x: "16GB" if x=="16GB" else "NA")
df["12GB"]=df["Ram"].apply(lambda x: "12GB" if x=="12GB" else "NA")
df["6GB"]=df["Ram"].apply(lambda x: "6GB" if x=="6GB" else "NA")
df["2GB"]=df["Ram"].apply(lambda x: "2GB" if x=="2GB" else "NA")
df["32GB"]=df["Ram"].apply(lambda x: "32GB" if x=="32GB" else "NA")
df["24GB"]=df["Ram"].apply(lambda x: "24GB" if x=="24GB" else "NA")

df

```
